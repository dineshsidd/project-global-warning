```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
sea_df = pd.read_csv("seaLevelData.csv")
sea_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(sea_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(8,4))
plt.bar(x_axis, sea_df["TOPEX/Poseidon"], color='b', alpha=0.5, align="edge")
plt.bar(x_axis, sea_df["Jason-1"], color='g', alpha=0.5, align="edge")
plt.bar(x_axis, sea_df["Jason-2"], color='y', alpha=0.5, align="edge")
plt.bar(x_axis, sea_df["Jason-3"], color='r', alpha=0.5, align="edge")
plt.title("Chance in Seal Level 1992 - 2018 (mm)")
plt.xlabel("Instance Recorded")
plt.ylabel("Change in Seal Level (mm)")
plt.tight_layout()
plt.savefig("seaLevelChange.png")
plt.show()
#plt.xticks(tick_locations, sea_df["year"], rotation="vertical")
```
